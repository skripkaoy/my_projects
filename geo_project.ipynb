# Выбор локации для скважины

**Цель исследования** — на основе предоставленных добывающей компанией «ГлавРосГосНефть» проб нефти в трёх регионах: в каждом 10 000 месторождений, где измерили качество нефти и объём её запасов, необходимо построить модель машинного обучения, которая поможет определить регион, где добыча принесёт наибольшую прибыль. Так же необходимо проанализировать возможную прибыль и риски техникой `Bootstrap`.

**Ход исследования**

В связи с тем, что о качестве данных ничего не известно, перед построением моделей необходимо провести обзор данных. После проверки данных на ошибки, различные артефакты, пропуски найдем возможность исправить самые критичные ошибки данных. Затем уже будем обучать модели на обучающей выборке и вычислять метрики на валидационной выборке.
 
Таким образом, исследование пройдёт в пять этапов:
 1. Обзор данных.
 2. Подготовка данных.
 3. Подготовка признаков к построению моделей.
 4. Обучение и проверка моделей.
 5. Расчёт прибыли и рисков.

**Описание данных:**

Данные геологоразведки трёх регионов находятся в файлах: 
- `/datasets/geo_data_0.csv`
- `/datasets/geo_data_1.csv`
- `/datasets/geo_data_2.csv`


- **Признаки**

`id` — уникальный идентификатор скважины

`f0`, `f1`, `f2` — три признака точек (неважно, что они означают, но сами признаки значимы)

- **Целевой признак**

`product` — объём запасов в скважине (тыс. баррелей)

**Шаги для выбора локации:**

- В избранном регионе ищут месторождения, для каждого определяют значения признаков;
- Строят модель и оценивают объём запасов;
- Выбирают месторождения с самым высокими оценками значений. Количество месторождений зависит от бюджета компании и стоимости разработки одной скважины;
- Прибыль равна суммарной прибыли отобранных месторождений.

**Условия задачи:**

- Для обучения модели подходит только линейная регрессия (остальные — недостаточно предсказуемые).
- При разведке региона исследуют 500 точек, из которых с помощью машинного обучения выбирают 200 лучших для разработки.
- Бюджет на разработку скважин в регионе — 10 млрд рублей.
- При нынешних ценах один баррель сырья приносит 450 рублей дохода. Доход с каждой единицы продукта составляет 450 тыс. рублей, поскольку объём указан в тысячах баррелей.
- После оценки рисков нужно оставить лишь те регионы, в которых вероятность убытков меньше 2.5%. Среди них выбирают регион с наибольшей средней прибылью.

Функция, выводящая все необходимые нам характеристики для данных.

```python
def dataframe_information(data):
    '''
    Функция принимает датафрейм для анализа и возвращает датафрейм с основными статистиками данных
    data - датафрейм
    Return:
    df_res - датафрейм с характеристиками данных
    '''
    
    df_data = []
    df_cols = ['name', 'object', 'na', 'zero', 'rate', 'unique', 'neg'
               , 'mean', 'moda', 'std', 'min', 'max', 'out left', 'out rigth', '25%', '50%', '75%']
    
    # перебираем столбцы в наборе данных
    for column_name in data.columns:
               
        # считаем характеристики
        column_type = data[column_name].dtypes
        column_na_values = data[column_name].isna().sum()
        column_zero_values = data[column_name][data[column_name] == 0].count()
        column_na_zero_rate = (column_na_values + column_zero_values) / data.shape[0]
        column_unique_values = data[column_name].nunique()
        column_negative_values = (data[column_name][data[column_name] < 0].count()
                                  if data[column_name].dtypes != 'object' else 0)
        column_mean = data[column_name].mean() if data[column_name].dtypes != 'object' else 0
        column_moda = statistics.mode(data[column_name])
        column_std = data[column_name].std() if data[column_name].dtypes != 'object' else 0
        column_min = data[column_name].min() if data[column_name].dtypes != 'object' else 0
        column_max = data[column_name].max() if data[column_name].dtypes != 'object' else 0
        outliers_left_number = (data[column_name][data[column_name] < data[column_name].quantile(.025)].count()
                                if data[column_name].dtypes != 'object' else 0)
        outliers_right_number = (data[column_name][data[column_name] > data[column_name].quantile(.975)].count()
                                 if data[column_name].dtypes != 'object' else 0)
        percentile_25 = data[column_name].quantile(.25) if data[column_name].dtypes != 'object' else 0
        percentile_50 = data[column_name].quantile(.50) if data[column_name].dtypes != 'object' else 0
        percentile_75 = data[column_name].quantile(.75) if data[column_name].dtypes != 'object' else 0
       
    # собираем показатели по каждому полю
        df_data.append([column_name, column_type, column_na_values
                                 , column_zero_values, column_na_zero_rate, column_unique_values
                                 , column_negative_values, column_mean, column_moda, column_std, column_min
                                 , column_max, outliers_left_number, outliers_right_number
                                 , percentile_25, percentile_50, percentile_75])
    
    # формируем набор данных
    df_res = pd.DataFrame(data = df_data, columns = df_cols)
    
    return df_res
```

Функция для поиска выбросов.

```python
def search_outliers(data, col, col_name=None):
    '''
    Функция принимает датафрейм и столбец, ищет выбросы с помощью IQR и строит boxplot
    data - датафрейм
    col - столбец датафрейма
    col_name - наименование столбца (для графика)
    Return:
    lower_fence - верхняя граница нормальных значений
    upper_fence - нижняя граница нормальных значений
    '''
    # находим 3-й и 1-й квартили
    Q3 = np.quantile(data[col], 0.75, axis=0)
    Q1 = np.quantile(data[col], 0.25, axis=0)
    # находим межквартильный размах
    IQR = Q3 - Q1
    # находим границы нормального размаха
    upper_fence = round(Q3 + 1.5 * IQR, 2)
    lower_fence = round(Q1 - 1.5 * IQR, 2)
    # выводим интервалы фактических значений и нормальных значений, количество аномальных значений
    print('-' * 45)
    print(f'Интервал фактических значений для переменной {col} от {round(data[col].min(),2)} до {round(data[col].max(),2)}')
    print(f'Интервал нормальных значений для переменной {col} от {lower_fence} до {upper_fence}')

    print(f'Переменная {col} содержит аномальных значений : {round(data[(data[col]>upper_fence)| (data[col]<lower_fence)].shape[0]/data.shape[0]*100, 2)}%')

    if data[(data[col] == 0)].shape[0]>0:
        print(f'Переменная {col} содержит нулевые значения {round(data[(data[col] == 0)].shape[0]/data.shape[0]*100, 2)}%')
 
    #  строим диаграмму размаха    
    if col_name != None:
        data.loc[:, [col]].boxplot().set(title=f'Диаграмма размаха значений переменной {col_name}',
                                         ylabel=col_name) #  строим диаграмму размаха
        plt.show()
    
    return lower_fence, upper_fence
```

Функция, обучающая модели линейной регрессии, делающая предсказания, а так же считающая средний запас предсказанного сырья и RMSE модели.

```python
# Метод обучения модели и прогнозирования
def lr_model(features_train, features_valid, target_train, target_valid, name_region,
             predicted_list, score_list, predicted_valid_mean_list):
    '''
    Функция обучает модель линейной регрессии, делает предсказания, считает средний запас предсказанного сырья и RMSE модели
    features_train - набор признаков обучающей выборки
    features_valid - набор признаков валидационной выборки
    target_train - целевой признак обучающей выборки
    target_valid - целевой признак валидационной выборки
    name_region - наименование региона
    predicted_list - список для предсказаний
    score_list - список для rmse модели
    predicted_valid_mean_list - список для значения среднего запаса предсказанного сырья
    '''
    
    # инициируем модели линейной регрессии
    model = LinearRegression()
    # обучим модель
    model.fit(features_train, target_train)
    # сделаем прогнозы
    predicted_valid = model.predict(features_valid) 
    # посчитаем rmse
    rmse = mean_squared_error(target_valid, predicted_valid)**0.5
    # посчитаем средний запас сырья
    predicted_valid_mean = predicted_valid.sum()/len(predicted_valid)
    
    # дополнение списка оценок, списка среднего запаса предсказанного сырья, сохранение предсказаний
    score_list.append(rmse)
    predicted_valid_mean_list.append(round(predicted_valid_mean, 6))
    predicted_list.extend(list(predicted_valid))
    
    print('#' * 36)
    print(f'## {name_region}. Результаты работы модели. ##')
    print('#' * 36)
    print()
    print(f'RMSE модели: {rmse}')
    print('Средний запас предсказанного сырья:', round(predicted_valid_mean, 6))
    print()
```

Функция для расчета средней прибыли, 95%-го доверительного интервала и риска убытков с помощью техники Bootstrap с 1000 выборок.

```python
def bootstrap_income(target, predicted):
    '''
    Функция принимает целевой признак и предсказания, и применя технику Bootstrap с 1000 выборок, 
    считает среднюю прибыль, 95%-й доверительный интервал и риск убытков.
    target - целевой признак
    predicted - предсказания
    Return:
    прибыль по выбранным скважинам
    '''
    income_200 = [] # список для расчитанной прибыли с выбранных скважин
    for i in tqdm(range(1000)):
        # в целевом признаке выбираем 500 точек для исследования
        target_subsample = target.reset_index(drop = True).sample(n = TOTAL_POINTS_REGION, 
                                                                  replace=True, 
                                                                  random_state=MY_RANDOM_STATE)
        predicted = pd.Series(predicted)
        # выбираем предсказания, соответствующие выбранным 500 точкам
        pred_subsumple = predicted[target_subsample.index] 
        # выбираем с помощью модели 200 самых прибыльных скважин и считаем по ним прибыль, добавляем в values
        income_200.append(income_selected_points(target_subsample, pred_subsumple))

    income_200 = pd.Series(income_200)
    # считаем среднее по посчитанным прибыли
    income_200_mean = int(income_200.mean())
    # вычисляем верхнююи нижнюю границы 95%-го доверительного интервала
    lower = int(income_200.quantile(q=0.025))
    upper = int(income_200.quantile(q=0.975))
    # вычисляем риск убытков
    risk = int(len(income_200[income_200 < 0]) / len(income_200) * 100)
    return income_200_mean, lower, upper, risk
```

## Подготовка к расчёту прибыли

Сохраним все ключевые значения для расчётов прибыли в отдельных переменных.

```python
# все ключевые значения для расчётов сохраним в отдельных переменных
TOTAL_POINTS_REGION = 500 # при разведке региона исследуют 500 точек
SELECTED_POINTS_REGION = 200 # с помощью машинного обучения выбирают 200 точек, лучших для разработки
REGION_BUDGET_RUB = 10*10**9 # бюджет на разработку скважин в регионе, руб
BARREL_INCOME_RUB = 450 # доход с одного барреля, руб
PRODUCT_INCOME_RUB = 450 * 10**3 # доход с каждой единицы продукта
LOSS_PROBABILITY = 2,5 # порог вероятности убытков
BUDGET_POINT_RUB = REGION_BUDGET_RUB / SELECTED_POINTS_REGION # бюджет на разработку одной скважины, руб
```

Рассчитаем достаточный объём сырья для безубыточной разработки новой скважины: для этого разделим бюджет для разработки одной скважины на доход с одной единицы продукта.

```python
# рассчитаем достаточный объём сырья для безубыточной разработки новой скважины
min_barrel_to_point = BUDGET_POINT_RUB / PRODUCT_INCOME_RUB
print(f'Достаточный объём сырья для безубыточной \
разработки новой скважины равен {int(min_barrel_to_point)} единиц продукта (тыс. баррелей)')
```
