# Исследование рынка компьютерных игр

**Цель исследования** — на основе исторических данных о продажах компьютерных игр, оценках пользователей и экспертов, жанрах и платформ (например, Xbox или PlayStation) интернет-магазина «Стримчик» выявить закономерности, определяющие успешность игры. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании. Необходимо будет проверить следующие гипотезы:
1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
2. Средние пользовательские рейтинги жанров Action и Sports разные.

**Ход исследования**

В связи с тем, что о качестве данных ничего не известно, перед исследованием данных необходимо провести обзор данных. После проверки данных на ошибки, различные артефакты найдем возможность исправить самые критичные ошибки данных.
 
Таким образом, исследование пройдёт в три этапа:
 1. Обзор данных.
 2. Предобработка данных.
 3. Исследовательский анализ данных.

**Описание данных:**

`Name` — название игры

`Platform` — платформа

`Year_of_Release` — год выпуска

`Genre` — жанр игры

`NA_sales` — продажи в Северной Америке (миллионы проданных копий)

`EU_sales` — продажи в Европе (миллионы проданных копий)

`JP_sales` — продажи в Японии (миллионы проданных копий)

`Other_sales` — продажи в других странах (миллионы проданных копий)

`Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Видим, что в столбце `Year_of_Release` хранится год, а тип данных - float64, необходимо изменить тип данных на datetime. Так же после обработки пропусков в столбцах `Critic_Score` и `User_Score` необходимо изменить тип данных на float.

Так же можно заметить, что названия столбцов не соответствуют общепринятым правилям - в названиях есть прописные буквы.

Напишем функцию, выводящую все необходимые нам характеристики для данных.

```python
# вывод основных статистик по набору данных

def dataframe_information(data):
    
    df_data = []
    df_cols = ['name', 'object', 'na', 'zero', 'rate', 'unique', 'neg'
               , 'mean', 'moda', 'std', 'min', 'max', 'out left', 'out rigth', '25%', '50%', '75%']
    
    # перебираем столбцы в наборе данных
    for column_name in data.columns:
               
        # считаем характеристики
        column_type = data[column_name].dtypes
        column_na_values = data[column_name].isna().sum()
        column_zero_values = data[column_name][data[column_name] == 0].count()
        column_na_zero_rate = (column_na_values + column_zero_values) / data.shape[0]
        column_unique_values = data[column_name].nunique()
        column_negative_values = (data[column_name][data[column_name] < 0].count()
                                  if data[column_name].dtypes != 'object' else 0)
        column_mean = data[column_name].mean() if data[column_name].dtypes != 'object' else 0
        column_moda = statistics.mode(data[column_name])
        column_std = data[column_name].std() if data[column_name].dtypes != 'object' else 0
        column_min = data[column_name].min() if data[column_name].dtypes != 'object' else 0
        column_max = data[column_name].max() if data[column_name].dtypes != 'object' else 0
        outliers_left_number = (data[column_name][data[column_name] < data[column_name].quantile(.025)].count()
                                if data[column_name].dtypes != 'object' else 0)
        outliers_right_number = (data[column_name][data[column_name] > data[column_name].quantile(.975)].count()
                                 if data[column_name].dtypes != 'object' else 0)
        percentile_25 = data[column_name].quantile(.25) if data[column_name].dtypes != 'object' else 0
        percentile_50 = data[column_name].quantile(.50) if data[column_name].dtypes != 'object' else 0
        percentile_75 = data[column_name].quantile(.75) if data[column_name].dtypes != 'object' else 0
       
    # собираем показатели по каждому полю
        df_data.append([column_name, column_type, column_na_values
                                 , column_zero_values, column_na_zero_rate, column_unique_values
                                 , column_negative_values, column_mean, column_moda, column_std, column_min
                                 , column_max, outliers_left_number, outliers_right_number
                                 , percentile_25, percentile_50, percentile_75])
    
    # формируем набор данных
    df_res = pd.DataFrame(data = df_data, columns = df_cols)
    
    return df_res
```

Выведим характеристики данных в датафрейме `data`.

```python
# вызываем функцию для расчета характеристик по данным
dataframe_information(data)
```

Посмотрим еще раз на количество пропусков.

Как мы видим, в данных строках помимо столбца `name` есть и другие пропуски - `genre`, `critic_score`, `user_score`, `rating` - как мы указывали выше, это важные для анализа столбцы, к тому же игры 1993 года. Возможно, именно из-за "возраста" игр о них так мало информации, или у игры не было утверждено официальное название и поэтому ячейка оказалась пуста. Можно предположить, что данные строки не сильно повлияют на результат исследования, и удалить их.

```python
# удаляем строки с пропусками данных в столбце name 
# делаем сброс индексов
data = data.dropna(subset=['name']).reset_index(drop=True)
```

Проверим пропуски в столбце `name`.

```python
# проверяем количество пропусков в таблице data
data.isnull().sum()
```

В столбце `name` больше нет пропусков.

- Следующий столбец с пропусками - `year_of_release`. Пропусков всего 2%. Пропуск в данном столбце мог возникнуть из-за того, что при заполнении данных об игре забыли указать дату релиза или точная дата выхода игры не определена. Посмотрим на эти строки.

```python
data.query('year_of_release.isna()')
```

Мы видим, что у некоторых игр год релиза есть в названии. Попробуем заполнить пропуски в столбце `year_of_release` на основании названия игры.

```python
idx_year = {}  # создаем пустой словарь
# итерируемся по строкам, в которых нет данных о году релиза игры
for row in data[data['year_of_release'].isna()].itertuples():
    try:
        if int(row.name[-4:]) > 1: # проверяем, что последние 4 символа в столбце name являются целым положительным числом
            if len(row.name[-4:]) == 4: # проверяем длину этого числа
                    item = int(row.name[-4:])  # преобразовываем тип данных
                    if item >= 1980:  # 1980 -- минимальное значение года в наших данных
                        idx_year[row.Index] = item # добавляем найденный год релиза и индекс строки в словарь
    except:  # если последние 4 символа не являются числом, то продолжаем итерировать по строкам
        continue
        
print(idx_year) # проверяем полученный словарь
```

Производим замену пропусков согласно получившемуся словарю.

```python
# итерирумся по словарю и производим замену
for key, year in idx_year.items():
    data.loc[key, 'year_of_release'] = year
```

Проверим оставшиеся пропуски в столбце `year_of_release`.

```python
data.query('year_of_release.isna()')
```

Пропусков стало чуть меньше. Так как их незначительное количество - меньше 2%, то можно предположить, что удаление этих данных не скажется на результатах исследования. Однако, если данных будет не хватать на последующих шагах, можно будет вернуться и оставить эти пропуски.

Удалим строки с пропущенными значениями в столбце `year_of_release`.

```python
# удаляем строки с пропусками данных в столбце year_of_release 
# делаем сброс индексов
data = data.dropna(subset=['year_of_release']).reset_index(drop=True)
```

Проверим пропуски в столбце `year_of_release`.

```python
# проверяем количество пропусков в таблице data
data.isnull().sum()
```

В столбце `year_of_release` все пропуски обработаны успешно.

- Обработаем пропуски в столбце `critic_score`. В этом столбце очень много пропусков - они составляют 51%. Посмотрим на эти строки.

```python
# выведим строки с пропусками данных в столбце critic_score
data[data['critic_score'].isna()]
```

### Влияние на продажи отзывов пользователей и критиков

Посмотрим, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Построим диаграммы рассеивания и посчитаем коэффициенты корреляции между отзывами и продажами. Для исследования возьмем самую популярную платформу - PS4.

Напишем функцию, которая будет строить диаграммы рассеяния по значениям отзывов критиков и продаж и по отзывам пользователей и продажам на платформе. А так же будет считать коэффициент корреляции Пирсона для глобальных продаж и оценки критиков, и глобальных продаж и оценки пользователей.

```python
def sales_critic_users_score(data_platform, platform):
    """
    Функция рисует диаграммы рассеяния по значениям отзывов критиков и продаж и  по отзывам критиков 
    и продажам на платформе. Так же считает коэффициент корреляции Пирсона глобальных продаж и оценки 
    критиков, и глобальных продаж и оценки пользователей.
    
    Parameters
    ----------
    data_platform: Pandas Dataframe
        Набор данных.
    platform: string
        По данным какой платформы строятся диаграммы.
    method: string
        count - значение по умолчанию, sum - второе возможное значение.
    sum_field: int
        Название поля, по которому будет посчитана сумма при группировке.
    n_items: string
        5 - значение по умолчанию. Количество возвращаемых значений.
    """

    # размечаем пространство для вывода графиков
    fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(18, 5))

    # построение диаграммы рассеивания по значениям отзывов критиков и продаж
    ax1.scatter(x=data_platform.critic_score, y=data_platform.all_sales, alpha=0.5)
    ax1.set_title('Диаграмма рассеивания по отзывам критиков и продажам на платформе ' + platform)
    ax1.set_xlabel('Отзывы критиков')
    ax1.set_ylabel('Количество проданных копий, млн')

    # построение и оформление графика рассеивания для отзывов пользователей
    ax2.scatter(x=data_platform.user_score, y=data_platform.all_sales, alpha=0.5)
    ax2.set_title('Диаграмма рассеивания по отзывам пользователей и продажам на платформе ' + platform)
    ax2.set_xlabel('Отзывы пользователей')
    ax2.set_ylabel('Количество проданных копий, млн')

    plt.show()
    # Подсчет и вывод на экран коэффициентов корреляции
    print(platform + ': ')
    print('Коэффициент корреляции глобальных продаж и оценки критиков:', 
      round(data_platform['all_sales'].corr(data_platform['critic_score']), 3))
    print('Коэффициент корреляции глобальных продаж и оценки пользователей:', 
      round(data_platform['all_sales'].corr(data_platform['user_score']), 3))
```

```python
# вызовем функцию sales_critic_users_score для набора данных по платформе PS4
sales_critic_users_score(data_2years.query('platform == "PS4"'), 'PS4')
```

Как видно по графику и коэффициенту корреляции Пирсона корреляция между продажами и оценками критиков на платформе PS4 положительная слабая (0.393). Корреляция между продажами и оценками пользователей очень слабая (-0.06), то есть связи нет.

## Составление портрета пользователя каждого региона

Составим для каждого региона портрет пользователя: определим наиболее популярные платформы, наиболее популярные жанры, влияние рейтинга ESRB на продажи по регионам, влияние оценок пользователей и критиков на продажи по регионам.

Напишем функцию, которая возвращает top-N значений столбца датафрейма. Все данные, которые не попали в top-N будут выделяться в категорию `other`.

```python
def top_n_items(df, group_field, method='count', sum_field='none', n_items=5):
    '''
    Функция возвращает top-N значений столбца датафрейма df, сгруппированных по полю group_field.
    
    Parameters
    ----------
    df: Pandas Dataframe
        Набор данных.
    group_field: string
        Поле, по которому будет сделана группировка данных.
    method: string
        count - значение по умолчанию, sum - второе возможное значение.
    sum_field: int
        Название поля, по которому будет посчитана сумма при группировке.
    n_items: string
        5 - значение по умолчанию. Количество возвращаемых значений.
    
    Returns
    -------
    top_N_items: list
        Cписок top-N значений поля group_field, полученного в результате применения метода method
        по полю sum_field к набору данных df
    df_grouped: Pandas Dataframe
        Датафрейм с top-N значений поля group_field и статистика после применения метода method
    '''
    
    # для подсчета количества группируем данные по полю group_field и считаем количество
    if method == 'count':
        df_grouped = df.groupby(group_field)['name'].count().reset_index()
    # для подсчета суммы группируем данные по полю group_field и считаем сумму по полю sum_field
    elif method == 'sum' and sum_field != 'none':    
        df_grouped = df.groupby(group_field)[sum_field].sum().reset_index()
    else:
        return
    
    df_grouped.columns = [group_field, method] # определяем наименование столбцов для df_grouped
    df_grouped.sort_values(by=method, ascending=False, inplace=True) # сортируем
    df_grouped_other_sum = float("{0:.2f}".format(df_grouped.iloc[5:][method].sum()))#считаем сумму для невошедших в топ
    df_grouped = df_grouped.head(n_items) # оставляем в датафрейме только топ-n
    df_grouped.loc[len(df_grouped.index)] = ['other', df_grouped_other_sum] #добавляем в датафрейм данные по категории other
    df_grouped.sort_values(method, ascending=False, inplace=True) # сортируем
    top_N_items = df_grouped[group_field].to_list() # определяем топ-n позиций
    
    return top_N_items, df_grouped
```

```python
def plot_region(df, sum_field, method='sum', n_items=5):
    
    """
    Функция выводит на экран набор графиков для проведения анализа портрета пользователя:
    - top-N платформ по продажам
    - top-N жанров по продажам
    - продажи по рейтингу
    - продажи top-N платформ по годам
    - диаграмма рассеяния critic_score - <region>_sales
    - диаграмма рассеяния user_score - <region>_sales
    
    Parameters
    ----------
    df: Pandas Dataframe
        Набор данных.
    sum_field: string
        Поле, по которому будет сделана группировка данных.
    method: string
        count - значение по умолчанию, sum - второе возможное значение.
    n_items: string
        5 - значение по умолчанию. Количество возвращаемых значений.
    """
    
    # получаем список и датафрейм top-N значений по платформе и жанру
    top_5_platforms, df_top_5_platforms = top_n_items(df, 'platform', method=method
                                                       , sum_field=sum_field, n_items=n_items) 
    top_5_genres , df_top_5_genres = top_n_items(df, 'genre', method=method
                                                 , sum_field=sum_field, n_items=n_items)
    
    # задаем 6 графиков
    fig, ax = plt.subplots(3,2, figsize=(20,20))
    fig.subplots_adjust(hspace=0.5)

    # top-N платформ по продажам
    ax[0][0].set_xlabel('Платформа', fontsize = 14)
    ax[0][0].set_ylabel('Продажи', fontsize = 14)
    ax[0][0].set_title('Продажи top-' + str(n_items) + ' платформ', fontsize = 14)
    ax[0][0].bar(df_top_5_platforms['platform'], df_top_5_platforms[method], color='orange')

    # top-N жанров по продажам
    ax[0][1].set_xlabel('Жанр', fontsize = 14)
    ax[0][1].set_ylabel('Продажи', fontsize = 14)
    ax[0][1].set_title('Продажи top-' + str(n_items) + ' жанров', fontsize = 14)
    ax[0][1].bar(df_top_5_genres['genre'], df_top_5_genres[method], color='orange')

    # продажи по рейтингу
    # группируем данные по рейтингу и считаем сумму в столбце с продажами по региону
    df_rating = df.groupby('rating')[sum_field].sum().reset_index()
    # определяем наименования столбцов
    df_rating.columns = ['rating', method]

    ax[1][0].set_xlabel('Рейтинг', fontsize = 14)
    ax[1][0].set_ylabel('Продажи', fontsize = 14)
    ax[1][0].set_title('Продажи от рейтинга', fontsize = 14)
    ax[1][0].bar(df_rating['rating'], df_rating[method], color='green')

    # продажи top-N платформ по годам
    # группируем данные по топ-n платформам по году релиза и платформе и считаем сумму продаж в регионе
    df_years = df.query('platform in @top_5_platforms').\
               groupby(['year_of_release','platform'])[sum_field].sum().reset_index()
    # определяем наименования столбцов
    df_years.columns = ['year_of_release', 'platform', method]
    # для каждой из платформ строим график по году релиза и продажам в этом регионе
    for platform in df_years['platform'].unique():
        ax[1][1].set_xlabel('Год', fontsize = 14)
        ax[1][1].set_ylabel('Продажи', fontsize = 14)
        ax[1][1].set_title('Продажи по годам', fontsize = 14)
        ax[1][1].plot(df_years[df_years['platform'] == platform]['year_of_release'], 
                      df_years[df_years['platform'] == platform][method], '--o', label=platform)
        ax[1][1].legend(loc='upper right')

    # диаграмма рассеяния critic_score - <region>_sales
    ax[2][0].set_xlabel('Отзывы критиков', fontsize = 14)
    ax[2][0].set_ylabel('Количество проданных копий, млн', fontsize = 14)
    # добавляем к названию графика коэффициент корреляции
    ax[2][0].set_title('Диаграмма рассеивания по отзывам критиков и продажам (' + 
                       str(round(df[sum_field].corr(df['critic_score']), 3)) + ')', fontsize = 14)
    ax[2][0].scatter(x=df['critic_score'], y=df[sum_field], alpha=0.5)

    # диаграмма рассеяния user_score - <region>_sales
    ax[2][1].set_xlabel('Отзывы пользователей', fontsize = 14)
    ax[2][1].set_ylabel('Количество проданных копий, млн', fontsize = 14)
    # добавляем к названию графика коэффициент корреляции
    ax[2][1].set_title('Диаграмма рассеивания по отзывам пользователей и продажам (' + 
                       str(round(df[sum_field].corr(df['user_score']), 3)) + ')', fontsize = 14)
    ax[2][1].scatter(x=df['user_score'], y=df[sum_field], alpha=0.5)

    plt.show()
```

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Проверим, являются ли средние пользовательские рейтинги платформ Xbox One и PC одинаковыми. Для проверки гипотезы необходимо сформулировать нулевую и альтернативную гипотезы. Нулевую гипотезу часто формулируют так, чтобы использовать знак равенства. То есть так как мы исследуем, одинаковы ли средние рейтинги платформ Xbox One и PC, то нулевая гипотеза будет предполагать, что равна. В качестве альтернативной гипотезы сформулируем двустороннюю, потому что она предполагает возможное отклонение и в большую, и в меньшую стороны, чсто нас и интересует. Таким образом, сформулируем нулевую и альтернативную гипотезы следующим образом:
1. Нулевая гипотеза: средние пользовательские рейтинги платформ Xbox One и PC равны.
2. Альтернативная гипотеза: средние пользовательские рейтинги платформ Xbox One и PC не равны.

В качестве статистического теста выберем "t-тест Стьюдента для независимых выборок" или "t-тест для двух независимых выборок".
T-тест Стьюдента для независимых выборок используется для определения статистической значимости различий между средними значениями двух независимых групп или выборок. Он основан на распределении Стьюдента и позволяет оценить, насколько значимы различия между средними значениями двух групп. 

Чтобы принять решение о том, использовать ли генеральную совокупность или взять случайную выборку, а так же, что указывать в параметре `equal_var`, посмотрим на размер наших данных.

Проверим за какое время у нас есть данные по этим двум платформам. 

```python
data_5years[data_5years['platform'] == 'XOne']['year_of_release'].min()
```

```python
data_5years[data_5years['platform'] == 'PC']['year_of_release'].min()
```

Видим, что периоды для двух платформ разные, поэтому возьмем информацию для обеих приставок с 2013 года.

```python
# создаем датафрейм с данными по платформе XOne
sample_platform_xone = data_5years[data_5years['platform'] == 'XOne']['user_score'].dropna()
```

```python
# создаем датафрейм с данными по платформе PC
sample_platform_pc = data_5years[(data_5years['platform'] == 'PC') & 
                                 (data_5years['year_of_release'] >= 2013)]['user_score'].dropna()
```

```python
# проверяем размерность датафрейма
sample_platform_xone.shape
```

```python
# проверяем размерность датафрейма
sample_platform_pc.shape
```

Размеры срезов не велики, в Python можно легко проанализировать и совокупность в несколько тысяч чисел, поэтому возьмем генеральные совокупности. Однако, так как размеры срезов данных не равны, укажем параметр `equal_var=False`.

Так как будем проверять двустороннюю гипотезу, в параметр `alternative` можно ничего не указывать, будет использовано значение по умолчанию - 'two-sided'.

Уровень значимости примем 0.05.

```python
# указываем нулевую и альтернативную гипотезы
h0 = 'Средние пользовательские рейтинги платформ Xbox One и PC равны'
h1 = 'Средние пользовательские рейтинги платформ Xbox One и PC не равны'
alpha = .05 # уровень значимости
# применяем t-критерий Стьюдента
results = st.ttest_ind(sample_platform_xone, 
                       sample_platform_pc,
                       equal_var=False)

# выводим гипотезы, уровень значимости
print(f'H0 (нулевая гипотеза): {h0}')
print(f'H1 (альтернативная гипотеза): {h1}')
print(f'Уровень значимости: {alpha}')
print('')

# выводим среднее с дисперсию по сэмплам
print('Sample_1')
print(f'Среднее значение: {sample_platform_xone.mean():.2f}, дисперсия: {sample_platform_xone.var():.2f}')
print('')
    
print('Sample_2')
print(f'Среднее значение: {sample_platform_pc.mean():.2f}, дисперсия: {sample_platform_pc.var():.2f}')
print('')
print(f'p-значение: {results.pvalue}')
print('')
    
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Мы видим, что средние значения пользовательских рейтингов на платформах Xbox One и PC не равны, осталось понять, достаточно ли большое различие мы наблюдаем с учётом разброса, чтобы отвергнуть нулевую гипотезу об их равенстве. 

P-значение (p-value) представляет собой вероятность получить такие или еще более экстремальные различия между выборками, если нулевая гипотеза (гипотеза о равенстве средних значений) верна. Чем меньше p-значение, тем более значимы различия между выборками. Мы получили значение p-value 14%, то есть различие между выборками не настолоько значимо, чтобы отвергнуть нулевую гипотезу, и  больше заданного уровня значимости - 5%. 

**Вывод:** нулевую гипотезу отвергнуть не удалось. Таким образом, у нас имеются сильные основания полагать, что средние показатели рейтинга платформы Xbox One и PC одинаковые.

### Средние пользовательские рейтинги жанров Action и Sports разные

Проверим, различаются ли средние пользовательские рейтинги жанров Action и Sports. Нулевую гипотезу опять же сформулируем так, чтобы использовать знак равенства. А в качестве альтернативной гипотезы сформулируем двустороннюю, потому что она предполагает возможное отклонение и в большую, и в меньшую стороны, чсто нас и интересует. Таким образом, сформулируем нулевую и альтернативную гипотезы следующим образом:
1. Нулевая гипотеза: средние пользовательские рейтинги жанров Action и Sports равны.
2. Альтернативная гипотеза: средние пользовательские рейтинги жанров Action и Sports не равны.

В качестве статистического теста выберем "t-тест Стьюдента для независимых выборок".

Чтобы принять решение о том, использовать ли генеральную совокупность или взять случайную выборку, а так же, что указывать в параметре equal_var, посмотрим на размер наших данных.

```python
# создаем датафрейм с данными по жанру Action
sample_genre_action = data_5years[data_5years['genre'] == 'Action']['user_score'].dropna()
```

```python
# создаем датафрейм с данными по жанру Sports
sample_genre_sports = data_5years[data_5years['genre'] == 'Sports']['user_score'].dropna()
```

```python
# проверяем размерность датафрейма
sample_genre_action.shape
```

```python
# проверяем размерность датафрейма
sample_genre_sports.shape
```

Датафреймы не большого размера, поэтому для проверки гипотезы будем использовать генеральные совокупности. Размер срезов данных различаются более, чем в 2 раза, поэтому укажем, что дисперсии не равны, и используем параметр `equal_var=False`.

Так как будем проверять двустороннюю гипотезу, в параметр `alternative` можно ничего не указывать, будет использовано значение по умолчанию - 'two-sided'.

Уровень значимости примем 0.05.

```python
# указываем нулевую и альтернативную гипотезы
h0 = 'Средние пользовательские рейтинги платформ Xbox One и PC равны'
h1 = 'Средние пользовательские рейтинги платформ Xbox One и PC не равны'
alpha = .05 # уровень значимости
# применяем t-критерий Стьюдента
results = st.ttest_ind(sample_genre_action, 
                       sample_genre_sports,
                       equal_var=False)

# выводим гипотезы, уровень значимости
print(f'H0 (нулевая гипотеза): {h0}')
print(f'H1 (альтернативная гипотеза): {h1}')
print(f'Уровень значимости: {alpha}')
print('')

# выводим среднее с дисперсию по сэмплам
print('Sample_1')
print(f'Среднее значение: {sample_genre_action.mean():.2f}, дисперсия: {sample_genre_action.var():.2f}')
print('')
    
print('Sample_2')
print(f'Среднее значение: {sample_genre_sports.mean():.2f}, дисперсия: {sample_genre_sports.var():.2f}')
print('')
print(f'p-значение: {results.pvalue}')
print('')
    
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Мы видим, что средние значения длительности поездки пользователей с подпиской и без подписки не равны, осталось понять, достаточно ли большое различие мы наблюдаем с учётом разброса, чтобы отвергнуть нулевую гипотезу об их равенстве. Мы получили значение p-value меньше 1% - это меньше заданного уровня значимости (даже если бы мы установили порог в 1%), то есть наблюдаемое значение очень маловероятно при такой нулевой гипотезе, и мы отвергаем её в пользу альтернативной.

**Вывод:** нам удалось отвергнуть нулевую гипотезу. Таким образом, у нас имеются сильные основания полагать, что средние пользовательские рейтинги жанров Action и Sports различаются.

**Выводы:** мы проверили две гипотезы о средних пользовательских рейтингах. В результате проверки гипотез, мы теперь можем с большой долей уверенности полагать, что:
